#### [ Scikit-Learn ML API 활용법 ]
- 기계학습 알고리즘을 python언어로 구현되어 있음
- 사용법이 모든 알고리즘 동일
	* 알고리즘 객체/인스턴스 생성
	* 학습 => 알고리즘객체변수명.fit(2D_데이터, 1D_타겟)
	* 평가 => 알고리즘객체변수명.score(2D_데이터, 1D_예측값)
	* 예측 => 알고리즘객체변수명.predict(2D_데이터) ---> 1D_예측값 반환
- 성능평가 
	* 분류/회귀 함수 => 함수명(1D_정답, 1D_예측값)

[1] 모듈로딩 및 데이터 준비 <hr>

```python
## [1-1] 모듈로딩
from sklearn.linear_model import LinearRegression					## PL 알고리즘 모듈
from sklearn.metrics import mean_squared_error, mean_absolute_error ## 성능평가 모듈
```

```python
## [1-2] 데이터 준비
## - 공부시간에 따른 점수 데이터 셋의 임의의 생성
hour = [[2], [4], [5], [7], [8], [9]]
jumsu = [30, 45, 67, 78, 79, 80]
```

[2] 데이터 분포 확인 <hr>

```python
## - 모듈 로딩
import matplotlib.pyplot as plt
```

```python
## - 데이터 분포
plt.plot(hour, jumsu, 'or')
plt.xlabel('Hour')
plt.ylabel('Jumsu')
plt.show()
```

[3] 학습 진행 <hr>
- 학습 방법 : 지도학습 => 데이터 + 타겟 / 회귀(Regression)로 수치 예측
- 학습 알고리즘 : 데이터 분포가 선형을 띄고 있어서 선형회귀로 진행

```python
## [3-1] 선형회귀 알고리즘 객체/인스턴스 생성
lr=LinearRegression()
```

```python
## [3-2] 학습진행
lr.fit(hour, jumsu)
```

```python
## ==> 선형식의 모델 파라미터 확인
print(f'lr.coef_: {lr.coef_}, lr.intercept_: {lr.intercept_}')

print(f'jumsu = {lr.coef_[0]}*hour + {lr.intercept_}')
```

[4] 성능 평가 <hr>

- 모델 성능 평가
- 손실 계산

```python
## [4-1] 모델 성능 평가 ==> 검증 데이터 
## - 결과범위 : 0.0 ~ 1.0
## - 결과분석 : 1.0에 가까울수록 성능 좋은 모델
lr.score(hour, jumsu)
```

```python
## [4-2] 손실 계산 ==> 검증 데이터
## - 결과범위 : 0 ~ 무한대
## - 결과분석 : 0에 가까울수록 좋은 모델

## - 모델이 결정한 선형식기반 결과 예측
pre_jumsu = lr.predict(hour)

## - 모델 예측값과 정답의 차이 계산
mse = mean_squared_error(jumsu, pre_jumsu)

mae = mean_absolute_error(jumsu, pre_jumsu)
```

```python
print(f'mse => {mse}, mae => {mae}')
```

[5] 예측 서비스 <hr>
- 웹 또는 웹에 포함된 모델에서 예측 서비스 진행
- 예측 메서드 : 알고리즘객체변수명.predict(2D_데이터) ==> 1D_예측값 반환
	* 전제조건 >> 학습 시 진행된 데이터 전처리과정을 거친 2D_데이터 전달

```python
new_data = [[24]]

pre_jumsu = lr.predict(new_data)

print(f'당신의 점수는 {pre_jumsu}점 입니다.')
```
